# ANN

```python
def neural_network_model(hidden1, hidden2):
    """
    Creating a neural network from two hidden layers
    using ReLU as activation function in the two hidden layers
    and a linear activation in the output layer.

    Parameters
    ----------
    hidden1 : int
        Number of neurons in first hidden layer.

    hidden2: int
        Number of neurons in second hidden layer.

    Returns
    -------
    model
        Fully connected neural network model with two hidden layers.
    """

    model = Sequential()
    # First hidden layer
    model.add(Dense(hidden1, activation="relu", name="layer1"))
    # Second hidden layer
    model.add(Dense(hidden2, activation="relu", name="layer2"))
    # Output layer
    model.add(Dense(1, activation="linear", name="layer3"))

    # Compile model
    model.compile(loss="mean_squared_error", optimizer="adam", metrics=["mse", "mae"])
    return model
```

```python
# Fit the model
model = neural_network_model(64, 32)
model.fit(
    np.array(list((x_train_rand))).astype(float),
    y_train_rand.values,
    epochs=50,
    batch_size=32,
    verbose=0,
)
y_pred_ANN_rand = model.predict(np.array(list((x_test_rand))))
mae_ANN_rand = mean_absolute_error(y_test_rand, y_pred_ANN_rand)
mse_ANN_rand = mean_squared_error(y_test_rand, y_pred_ANN_rand)
r2_ANN_rand = r2_score(y_test_rand, y_pred_ANN_rand)
pearson_ANN_rand = np.corrcoef(y_pred_ANN_rand.flatten(), y_test_rand.values.flatten())[0, 1]
spearman_ANN_rand = spearmanr(y_pred_ANN_rand.flatten(), y_test_rand.values.flatten()).correlation
```
